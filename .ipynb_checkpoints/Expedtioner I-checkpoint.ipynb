```python
import pandas as pd
import numpy as np

loc = 'exped.xls'
df = pd.read_excel(loc)
```

```python
df.head
```

```python
df.WHEN.dtype
```

```python
# Use .dt.month to call just the months
df["QUARTER"] = df["WHEN"].dt.quarter
df['MONTH'] = df["WHEN"].dt.month
df['YEAR'] = df['WHEN'].dt.year
df['DAY'] = df['WHEN'].dt.day
df.head()
```

#### a) What was the value of catalog sales for London in the first quarter?

```python
df[(df["WHERE"]=="London") & (df["HOW"]=="Catalog") & (df['QUARTER']==1)]
```

```python
pd.pivot_table(df[(df["WHERE"]=="London") & (df["HOW"]=="Catalog") & (df['QUARTER']==1)], 
               index = ["WHERE", "HOW", 'QUARTER'],
               values = 'REVENUE', 
               aggfunc = np.sum, 
               margins=True, 
               margins_name = 'Total')
```

#### b) What percent of the total annual sales were Tokyo Web sales in the fourth quarter?

```python
pd.pivot_table(df[(df['WHERE']=="Tokyo") & (df["HOW"]=="Web") & (df["QUARTER"]==4)],
               index = ["WHERE", "HOW", 'QUARTER'],
               values = 'REVENUE',
               aggfunc = np.sum,
               margins=True, margins_name = 'Total')
```

#### d) What was the value of catalog sales for London in January and give details of the transactions?

```python
df[(df["WHERE"]=="London") & (df["HOW"]=="Catalog")&(df["MONTH"]==1)]
```

```python
pd.pivot_table(df[(df["WHERE"]=="London")&(df["HOW"]=="Catalog")&(df["MONTH"]==1)],
              index=["WHERE", "HOW", "MONTH"],
              values = "REVENUE",
              aggfunc = np.sum,
              margins = True,
              margins_name = "Total")
```

#### e) What was the value of Camel saddle sales for Paris in 2002 by quarter?

```python
pd.pivot_table(df[(df['WHERE']=="Paris") & (df['WHAT']=="Camel saddle")],
                index = ['WHAT',"QUARTER"], 
                values = "REVENUE",
                aggfunc = np.sum, margins = True, margins_name = "Total" )
```

#### f) How many Elephant polo sticks were sold in New York in each month of 2002?

```python
pd.pivot_table(df[(df['WHERE']=="New York") & (df['WHAT']=="Elephant polo stick")],
                index = ['WHAT',"MONTH"],
                values = "QUANTITY",
                aggfunc = np.sum, margins = True, margins_name = "Total" )
```
